#### Standardize the data

```python
c=df.columns
for i in c:
    df[i]=df[i].replace('Not Given',pd.NA)
    print(i,'OK')
```

##### since there are 2588 missing values and most of the other values are 'Not Given' in the column director, we can drop that column

#### Fill the country null value with the mode of the year

```python
df['country']=df.country.fillna(df['country'].mode()[0])
```

```python
df.isnull().sum()
```

#### Checking for outliers in duration column

#### Function to extract integers from duration string

```python
def extract_duration(duration):
    try:
        return int(''.join(filter(str.isdigit, duration)))
    except:
        return None
```

```python
df['duration'] = df['duration'].apply(extract_duration)
```

```python
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
outliers = df[(df['duration'] < lower_bound) | (df['duration'] > upper_bound)]
len(outliers)
```

##### Remove the outliers from the duration column

```python
df_cleaned_duration = df[~((df['duration'] < lower_bound) | (df['duration'] > upper_bound))]
```

#### Checking for outliers in release year column

```python
Q1_year = df['release_year'].quantile(0.25)
Q3_year = df['release_year'].quantile(0.75)
IQR_year = Q3_year - Q1_year
lower_bound_year = Q1_year - 1.5 * IQR_year
upper_bound_year = Q3_year + 1.5 * IQR_year
```

```python
outliers_year = df[(df['release_year'] < lower_bound_year) | (df['release_year'] > upper_bound_year)]
outliers_year
```

#### Removing outliers from the release year column

```python
df_cleaned = df_cleaned_duration[~((df_cleaned_duration['release_year'] < lower_bound_year) | (df_cleaned_duration['release_year'] > upper_bound_year))]
```

#### Resetting index after removing the outliers

```python
df_cleaned.reset_index(drop=True, inplace=True)
```
