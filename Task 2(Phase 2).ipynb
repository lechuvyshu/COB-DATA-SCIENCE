#### Import the required libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

#### Loading the train data

```python
train=pd.read_html("https://docs.google.com/spreadsheets/u/1/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml?urp=gmail_link")
train
```

```python
trainset=train[0]
trainset.to_csv('train.csv', index=False)
```

```python
train=pd.read_csv("train.csv")
```

```python
train
```

#### Making the train data more concise

```python
trains=train.drop('Unnamed: 0',axis=1)
trains.columns=trains.iloc[0]
train_data=trains.drop(0)
train_data
```

#### Loading the test data

```python
test=pd.read_html("https://docs.google.com/spreadsheets/u/1/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pubhtml")
test
```

```python
testset=test[0]
testset.to_csv('test.csv', index=False)
```

```python
test=pd.read_csv("test.csv")
test
```

#### Making the test data more concise

```python
tests=test.drop('Unnamed: 0',axis=1)
tests.columns=tests.iloc[0]
test_data=tests.drop(0)
test_data
```

#### Dropping the null rows in column y

```python
train_dataset=train_data.dropna(subset=['y'])
train_dataset
```

#### Distinguishing x_train and y_train

```python
x_train=train_dataset[['x']].astype(float)
y_train=train_dataset[['y']].astype(float)
```

```python
x_train
```

```python
y_train
```

#### Distinguishing x_test and y_test

```python
x_test=test_data[['x']].astype(float)
y_test=test_data[['y']].astype(float)
```

```python
x_test
```

```python
y_test
```

#### Linear Regression model for train

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

#### Predicting the test data

```python
y_pred=model.predict(x_test)
y_pred
```

#### Converting arrays to 1D 

```python
y_test=y_test.squeeze()
y_pred=y_pred.squeeze()
```

#### Prediction

```python
result=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred})
result
```

#### Evaluating the model using mean_squared_error and r2_score

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

#### Conclusion

##### A lower MSE indicates that the model's predictions are closer to the actual values, which suggests better predictive accuracy.

##### Our R2 approximately equal to 1 means that the model perfectly explains all the variance in the target variable.
